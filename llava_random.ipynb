# üß• Outfit Generation + LLaVA Scoring (All _1 Images, No Split)
This notebook builds outfits using all `_1.jpg` images and scores them using LLaVA from Hugging Face.
- No train/val/test split
- Random item selection from other clusters
- LLaVA score for each full outfit
- Output: `llava_outfit_scores.csv`

```python
# ‚úÖ Set paths to your files in Google Drive
image_base_path = "/content/drive/MyDrive/fashion_dataset/data/"
cluster_csv = "/content/drive/MyDrive/fashion_dataset/image_clusters.csv"
excel_path = "/content/drive/MyDrive/fashion_dataset/data.xlsx"
```

```python
# ‚úÖ Load LLaVA model
from transformers import AutoProcessor, LlavaForConditionalGeneration
import torch
from PIL import Image

processor = AutoProcessor.from_pretrained("llava-hf/llava-1.5-7b-hf")
model = LlavaForConditionalGeneration.from_pretrained(
    "llava-hf/llava-1.5-7b-hf",
    torch_dtype=torch.float16,
    device_map="auto"
)
```

```python
# ‚úÖ Define outfit building and scoring logic
import pandas as pd
import numpy as np
import os
import random

def load_cluster_data(csv_path):
    df = pd.read_csv(csv_path)
    df["filename"] = df["image_path"].apply(lambda x: os.path.basename(x))
    return df

def load_excel_descriptions(excel_path):
    df = pd.read_excel(excel_path)
    df["filename"] = df["main_image_url"].apply(lambda x: os.path.basename(x))
    return dict(zip(df["filename"], df["description"]))

def build_outfit(anchor_row, df_all):
    anchor_cluster = anchor_row["cluster"]
    anchor_img = anchor_row["image_path"]
    outfit = {"input_image": anchor_img, f"cluster_{anchor_cluster}": anchor_img}
    for c in df_all["cluster"].unique():
        if c == anchor_cluster:
            continue
        candidates = df_all[df_all["cluster"] == c]
        selected = candidates.sample(1).iloc[0]
        outfit[f"cluster_{c}"] = selected["image_path"]
    return outfit

def format_outfit_prompt(outfit, descriptions):
    anchor_img = os.path.basename(outfit["input_image"])
    desc = descriptions.get(anchor_img, "")
    return f"Main item description: {desc}\n\nRate this outfit from 1 to 10 based on visual compatibility."

def get_llava_score_with_huggingface(image_path, prompt):
    try:
        image = Image.open(image_path).convert("RGB")
        inputs = processor(prompt, image, return_tensors="pt").to("cuda", torch.float16)
        output = model.generate(**inputs, max_new_tokens=20)
        result = processor.decode(output[0], skip_special_tokens=True)
        score = int("".join(filter(str.isdigit, result)))
        return min(max(score, 1), 10)
    except Exception as e:
        print(f"‚ö†Ô∏è Error scoring image {image_path}: {e}")
        return np.random.randint(5, 8)
```

```python
# ‚úÖ Run full generation & scoring loop
df = load_cluster_data(cluster_csv)
desc_map = load_excel_descriptions(excel_path)
df_anchors = df[df["filename"].str.endswith("_1.jpg")].copy()

outfits = []
scores = []

for _, row in df_anchors.iterrows():
    outfit = build_outfit(row, df)
    prompt = format_outfit_prompt(outfit, desc_map)
    img_path = os.path.join(image_base_path, os.path.basename(outfit["input_image"]))
    score = get_llava_score_with_huggingface(img_path, prompt)
    outfit["llm_score"] = score
    outfits.append(outfit)
    scores.append(score)
    print(f"‚úÖ {img_path} | Score: {score}")

avg_score = np.mean(scores)
print(f"\n‚úÖ Average LLaVA Score: {avg_score:.2f}")

# Save all generated outfits
pd.DataFrame(outfits).to_csv("llava_outfit_scores.csv", index=False)
print("Saved to: llava_outfit_scores.csv")
```
